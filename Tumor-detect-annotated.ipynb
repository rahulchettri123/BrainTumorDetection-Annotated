```python
import os

base = '/kaggle/working/dataset'
os.makedirs(f'{base}/images/train', exist_ok=True)
os.makedirs(f'{base}/images/val', exist_ok=True)
os.makedirs(f'{base}/labels/train', exist_ok=True)
os.makedirs(f'{base}/labels/val', exist_ok=True)
```

```python
import shutil
import glob

# Copy training images
for folder in glob.glob('/kaggle/input/brain-tumor-scans-annotated/Brain Tumor Annotated/Images/Training/*'):
    for img in glob.glob(f'{folder}/*.jpg'):
        shutil.copy(img, f'{base}/images/train/')

# Copy testing images
for folder in glob.glob('/kaggle/input/brain-tumor-scans-annotated/Brain Tumor Annotated/Images/Testing/*'):
    for img in glob.glob(f'{folder}/*.jpg'):
        shutil.copy(img, f'{base}/images/val/')

# Copy training labels
for txt in glob.glob('/kaggle/input/brain-tumor-scans-annotated/Brain Tumor Annotated/Labels/Training/*.txt'):
    shutil.copy(txt, f'{base}/labels/train/')

# Copy testing labels
for txt in glob.glob('/kaggle/input/brain-tumor-scans-annotated/Brain Tumor Annotated/Labels/Testing/*.txt'):
    shutil.copy(txt, f'{base}/labels/val/')
```

```python
data_yaml = """
train: /kaggle/working/dataset/images/train
val: /kaggle/working/dataset/images/val

nc: 4
names: ['glioma', 'meningioma', 'pituitary', 'notumor']
"""

with open('/kaggle/working/dataset/data.yaml', 'w') as f:
    f.write(data_yaml)
```
